```python
import spacy
nlp = spacy.load('en_core_web_sm')
from spacy import displacy
```

```python
with open('/home/ravik/src_git/learn_NLP/S3/story_assesment.txt', encoding='utf-8') as f:
     doc = nlp(f.read())
```

```python
# all created tokens.
for token in doc:
    print(token.text)
```

```python
# Get total tokens number
total_tokens = 0
for token in doc:
    total_tokens += 1
print(total_tokens)
```
